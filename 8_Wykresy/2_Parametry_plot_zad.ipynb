```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
import datetime
nasa=pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/nasa_facebook_statuses.csv', 
                   usecols=["status_published","num_reactions","num_comments","num_shares","num_likes"])
nasa["status_published"] = pd.to_datetime(nasa["status_published"])
filter =(nasa['status_published'] >= '2016-5-1') & (nasa['status_published'] < '2016-6-1')
nasa = nasa[filter]
nasa['day'] = nasa.apply(lambda row: row["status_published"].day,axis=1)
nasa_by_day = nasa.groupby(by='day').sum()
nasa_by_day.head()
```

```python
nasa_by_day.plot()
```

```python
nasa_by_day.plot(figsize=(15, 5))
```

```python
nasa_by_day.plot(figsize=(15, 5), title='NASA Facebook activity')
```

```python
nasa_by_day.plot(figsize=(15, 5), title='NASA Facebook activity', subplots=True)
```

```python
nasa_by_day.plot(figsize=(15, 5), title='NASA Facebook activity', logy=True)
```

```python
nasa_by_day.plot(figsize=(15, 5), title='NASA Facebook activity', xticks=(range(32)), rot=45)
```
