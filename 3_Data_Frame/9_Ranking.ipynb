**Ranking**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
frame = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/Mcdonalds.csv',
                   usecols=['Item', 'Category', 'Serving Size', 'Calories', 'TotalFat'])
frame.head()
```

Wyznaczanie rankingu:

```python
frame['Calories'].rank().head()
```

Dodawanie kolumny z rankingiem do tabeli:

```python
frame['Calories Rank'] = frame['Calories'].rank()
frame.head()
```

```python
frame.sort_values(by='Calories').head()
```

rank(ascending=False) - wyznaczanie rankingu malejąco

```python
frame['Calories Rank'] = frame['Calories'].rank(ascending=False)
frame.head()
```

```python
frame.sort_values(by='Calories', ascending=False).head()
```

Metoda do wyciągania największych wartości z tabeli - nlargest:

```python
frame.sort_values(by='Calories', ascending=False, inplace=True)
frame.head(3)
```

```python
frame.nlargest(n=3, columns='Calories')
```
