**Oczyszczanie danych - metoda fillna()**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
frame = pd.read_csv("C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/Mcdonalds.csv")
frame.head()
```

```python
frame.loc[1, "TotalFat"] = np.NaN
frame.loc[1, "Saturated Fat"] = np.NaN
frame.loc[2, "Saturated Fat"] = np.NaN
frame.loc[3, "Category"] = np.NaN
frame.head()
```

fillna() - uzupeÅ‚nienie brakujÄ…cych danych, gdzie value=0

```python
frame.fillna(value=0).head()
```

UzywajÄ…c sÅ‚ownika moÅ¼na zamieinÄ‡ poszczegÃ³lne dane, np. kategoriÄ™

```python
replace_table = {"Category": "Unknown", "TotalFat": 0, "Saturated Fat": 0}
replace_table
```

```python
frame.fillna(value=replace_table).head()
```

Zmiana wartoÅ›ci tylko wybranej kolumny:

```python
frame["Category"].fillna(value="Unknown", inplace=True)
frame.head()
```

```python
frame["TotalFat"].fillna(value=0, inplace=True)
frame.head()
```

```python
frame["Saturated Fat"].fillna(value=0, inplace=True)
frame.head()
```

```python
frame.loc[1, "TotalFat"] = np.NaN
frame.loc[1, "Saturated Fat"] = np.NaN
frame.loc[2, "Saturated Fat"] = np.NaN
frame.loc[3, "Category"] = np.NaN
frame.head()
```

Usatwienie NaN jako Å›redniÄ… caÅ‚ej kolumny:

```python
avg_total_fat = frame["TotalFat"].mean()
avg_total_fat
```

```python
frame["TotalFat"].fillna(value=avg_total_fat, inplace=True)
frame.head()
```

fillna(method="ffill") zamienia NaN na wartoÅ›Ä‡ z wiersza powyÅ¼ej

```python
frame["Saturated Fat"].fillna(method="ffill", inplace=True)
frame.head()
```
