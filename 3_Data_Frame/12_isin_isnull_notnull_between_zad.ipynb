```python
import pandas as pd
import numpy as np
```

```python
fuel =pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/fuel.csv',low_memory=False,
        usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)','Start Stop Technology'])
fuel.head()
```

```python
list_of_cars = ['Renault', 'Toyota','Ford']
isin_list = fuel['Make'].isin(list_of_cars)
```

```python
fuel[isin_list].head()
```

```python
no_start_stop = fuel['Start Stop Technology'].isnull()
fuel[no_start_stop].head()
```

```python
is_start_stop = fuel['Start Stop Technology'].notnull()
fuel[is_start_stop].head()
```

```python
mpg50_60 = fuel['Combined MPG (FT1)'].between(50, 60)
fuel[mpg50_60].head()
```

```python
print(f'Jest {len(fuel[isin_list])} aut z listy: {list_of_cars}')
print(f'Jest {len(fuel[no_start_stop])} aut bez start/stop')
print(f'Jest {len(fuel[is_start_stop])} aut z start/stop')
print(f'Jest {len(fuel[mpg50_60])} aut z MPG 50 - 60')
```
