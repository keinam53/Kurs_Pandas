**Oczyszczanie danych - metoda dropna()**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
frame = pd.read_csv("C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/Mcdonalds.csv")
frame.head()
```

W tabeli podane są wszystkie wartości - wprowadzone zostaną wartości NaN

```python
frame.loc[1, "Calories"] = np.NaN
frame.loc[2, "TotalFat"] = np.NaN
```

```python
frame.head()
```

dropna() usuwa cały wiersz w którym jest NaN

```python
frame.dropna().head()
```

dropna(how="all") usuwa wiersze w których wszystkie wartości to NaN

```python
frame.dropna(how="all").head()
```

dropna(subset=[" "]) usuwa wiersz jeśli w podanej kolumnie jest NaN

```python
frame.dropna(subset=["TotalFat"], how="any").head()
```

dropna(axis="rows") - usuwa wiersze z NaN, dropna(axis="columns") - kolumny

```python
frame.dropna(axis="rows").head()
```

```python
frame.dropna(axis="columns").head()
```

dropna(inplace=True) - zapisuje zminę, False tworzy nowy obiekt

```python
frame.dropna(inplace=True)
frame.head()
```
