**Metody where i query**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
frame = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/Mcdonalds.csv',
                   usecols=['Item', 'Category', 'Serving Size', 'Calories', 'TotalFat'])
frame.head()
```

```python
calories_greater_400 = frame['Calories'] >= 400
calories_greater_400[:5]
```

Filtrowanie przy użyciu metody where()

where() dla wartości False ustawia NaN, zwykłe filtrowanie usuwa te kolumny

```python
frame.where(calories_greater_400).head()
```

```python
frame[calories_greater_400].head()
```

Można to samo osiągnąć używając metody dropna()

```python
frame.where(calories_greater_400).dropna(how='all').head()
```

query() - podobnie jak w SQL

```python
frame.query('Category == "Desserts"').head()
```

Można łączyć zapytania:

```python
frame.query('Category == ["Desserts", "Beverages"]').head()
```

```python
frame.query('Category == "Desserts" and Calories < 200').head()
```
