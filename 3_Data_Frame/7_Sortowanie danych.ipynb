**Sortowanie danych w Data Frame**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
frame = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/Mcdonalds.csv',
                   usecols=['Item', 'Category', 'Serving Size', 'Calories', 'TotalFat'])
frame.head()
```

sort_values(by=' ') - sortowanie wybranej kolumny od najmniejszych

```python
frame.sort_values(by='Calories').head()
```

ascending=False - od najwiÄ™kszych

```python
frame.sort_values(by='Calories', ascending=False).head()
```

DomyÅ›lnie NaN sÄ… wartoÅ›ciami najwiÄ™kszymi

```python
frame['Calories'] = frame['Calories'].astype('float')
frame.loc[82, 'Calories'] = np.NaN
frame.sort_values(by='Calories').tail()
```

na_position='first' - jako najmniejsze

```python
frame.sort_values(by='Calories', na_position='first').head()
```

MoÅ¼na sortowaÄ‡ ze wzglÄ™du na kilka kolumn:

```python
frame.sort_values(by=['Category', 'Item']).head(20)
```

Sortowanie kategorii rosnÄ…co a nazwy malejÄ…co:

```python
frame.sort_values(by=['Category', 'Item'], ascending=[True, False]).head(20)
```
