```python
kolor = ['zielony', 'zielony', 'niebieski', 'niebieski', 'czerwony', 'czerwony', 'biały']
rozmiar = ['S', 'S', 'S', 'M', 'M', 'M', 'L']
płeć = ['M', 'K', 'M', 'K', 'M', 'K', 'U']
clothes = pd.DataFrame({'kolor': kolor, 'rozmiar': rozmiar, 'płeć': płeć})
clothes
```

```python
is_color_blue = clothes['kolor'] == 'niebieski'
is_color_green = clothes['kolor'] == 'zielony'
is_color_red = clothes['kolor'] == 'czerwony'
```

```python
is_color_blue
```

```python
clothes[is_color_blue | is_color_green | is_color_red]
```

To samo co powyżej za pomocą metody isin()

```python
moje_kolory = clothes['kolor'].isin(['zielony', 'niebieski', 'czerwony'])
clothes[moje_kolory]
```

```python
clothes.loc[6, 'płeć'] = np.NaN
clothes
```

Metoda notnull()

```python
has_gender = clothes['płeć'].notnull()
clothes[has_gender]
```

Metoda notnull()

```python
gender_unknown = clothes['płeć'].isnull()
clothes[gender_unknown]
```
