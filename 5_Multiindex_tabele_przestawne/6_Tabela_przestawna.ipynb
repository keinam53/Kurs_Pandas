**Pivot table - tabela przestawna**

narzędzie analityczne pozwalające filtrować, wybierać, przestawiać kolumny i wiersze z danymi w arkuszu kalkulacyjnym, aby uzyskać wynikowy, odmienny, bardziej przejrzysty i użyteczny od pierwotnego (źródłowego) układ danych, bez naruszania oryginalnej źródłowej tabeli lub tabel.

```python
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/WA_Sales_Products_2012-14.csv')
sales.head()
```

Tworzenie tabeli przestawnej

```python
sales.pivot_table(index='Retailer country', columns=['Year', 'Quarter'], values='Revenue').head()
```

Zestawienie średnich rocznych zysków

```python
sales.pivot_table(index='Retailer country', columns='Year', values='Revenue').head()
```

Zestawienie z minimalną marżą

```python
sales.pivot_table(index='Retailer country', columns='Year', values='Gross margin', aggfunc='min').head()
```

Można użyć 2 lub więcej funkcji

```python
sales.pivot_table(index='Retailer country', columns='Year', values='Gross margin', aggfunc=['min', 'max']).head()
```

```python
piv_tab = sales.pivot_table(index='Retailer country', columns='Year', values='Gross margin', aggfunc=['min', 'max'])
piv_tab.columns
```

```python
piv_tab = piv_tab.swaplevel(axis='columns').sort_index(axis='columns')
piv_tab.head()
```
