**Swaplevel**

```python
import pandas as pd
import numpy as np
```

```python
incidents = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/Canadian Railway Crossing Incidents.csv')
incidents.set_index(['Region', 'EventType'], inplace=True)
incidents.sort_index(inplace=True)
incidents.head(10)
```

Przed zmianą indeksu należy go zresetować

```python
incidents.reset_index(inplace=True)
incidents.set_index(['EventType', 'Region'], inplace=True)
incidents.sort_index(inplace=True)
incidents.head()
```

Można to samo zrobić metodą swaplevel()

```python
incidents.swaplevel().head()
```

```python
incidents.swaplevel().sort_index().head()
```

```python
incidents = incidents.swaplevel().sort_index()
incidents
```
