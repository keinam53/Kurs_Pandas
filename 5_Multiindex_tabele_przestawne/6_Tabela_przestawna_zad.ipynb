```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import time
```

```python
df = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/marathon_results_2016.csv',
                index_col='Bib', usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
df.pivot_table(index='Age', columns='M/F', values='TotalSeconds').head()
```

```python
df.pivot_table(index='Age', columns='M/F', values=['TotalSeconds', 'HalfSeconds']).head()
```

```python
df.pivot_table(index="Age",columns="M/F",values=["HalfSeconds","TotalSeconds"]).swaplevel(axis=1).head()
```

```python
df.pivot_table(index="Age",columns="M/F",values=["HalfSeconds","TotalSeconds"]).swaplevel(axis=1).sort_index(axis=1).head()
```

```python
df.pivot_table(index='Age', columns='M/F', values='TotalSeconds', aggfunc='count').head()
```

```python
df.pivot_table(index='Age', values='TotalSeconds', aggfunc='count').head()
```

```python
df.pivot_table(index='Country', values='TotalSeconds', aggfunc='count').head()
```

```python
df.pivot_table(index='Country', values='TotalSeconds', aggfunc='count').sort_values(by='TotalSeconds',
                ascending=False).head()
```

```python
df.pivot_table(index=['Country', 'State', 'City'], values='TotalSeconds').head()
```

```python
df.pivot_table(index=['Country', 'State', 'City'], values='TotalSeconds').sort_values(by='TotalSeconds',
                    ascending=False).head()
```
