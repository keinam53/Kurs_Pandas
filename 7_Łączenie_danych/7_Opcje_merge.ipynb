**Opcje polecenia merge**

```python
import pandas as pd
import numpy as np
```

```python
products = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/northwind/products.csv',
                      usecols=['ProductID', 'ProductName', 'CategoryID', 'UnitPrice'])
categories = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/northwind/categories1.csv',
                        usecols=['CategoryID', 'CategoryName', 'Description'])
products.head()
```

```python
categories.head()
```

```python
products.merge(right=categories, on='CategoryID').head()
```

Zmiana nazwy kolumny

```python
categories.rename({'CategoryID': 'ID'}, axis='columns', inplace=True)
categories.head()
```

Jeśli są inne nazwy kolumn należy je zmienić lub skorzystać z atrybutów left_on i right_on

```python
products.merge(right=categories, left_on='CategoryID', right_on='ID').head()
```

Jeśli w innych komumnach są takie same nazwy - polecenie dzaiała poprawnie, można użyć suffixes

```python
products.rename({'ProductName': 'Name'}, axis='columns', inplace=True)
categories.rename({'CategoryName': 'Name'}, axis='columns', inplace=True)
products.merge(right=categories, left_on='CategoryID', right_on='ID', suffixes=['_prod', '_cat']).head()
```

Można posortować w oparciu o wartość, która była kluczem łączenia

```python
products.merge(right=categories, left_on='CategoryID', right_on='ID', suffixes=['_prod', '_cat'], sort=True).head()
```

Łączenie w oparciu o indeks

```python
categories.set_index('ID', inplace=True)
categories.head()
```

```python
products.merge(right=categories, left_on='CategoryID', right_index=True, suffixes=['_prod', '_cat'], sort=True).head()
```
