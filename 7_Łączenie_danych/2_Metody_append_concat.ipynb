Zmiana nazwy kolumny tak zeby była taka sama w obu tabelach

```python
suppliers2 = suppliers.copy()
suppliers2.rename({'SupplierID': 'CustomerID'}, axis='columns', inplace=True)
suppliers2.head()
```

Po ponownym złączeniu jest już jedna kolumna

```python
df_concat = pd.concat(objs=[suppliers, customers], ignore_index=True)
len(df_concat)
```
