**Metoda merge**

```python
import pandas as pd
import numpy as np
```

```python
cat_1 = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/northwind/categories_del_1.csv',
                    usecols=['CategoryID', 'CategoryName'])
cat_2 = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/northwind/categories_del_2.csv',
                    usecols=['CategoryID', 'CategoryName'])
```

Połączenie części wspólnej

```python
cat_1.merge(cat_2)
```

```python
cat_1.merge(cat_2, on='CategoryID')
```

```python
cat_1.merge(cat_2, on='CategoryID', suffixes=['_1', '_2'])
```

Parametr how= inner(część wspólna), left(lewy zbiór), right(prawy zbiór), outer(lewy i prawy)

```python
cat_1.merge(cat_2, on='CategoryID', suffixes=['_1', '_2'], how='outer')
```

indicatro - pokazuje w którym zbiorze była wartość

```python
cat_merge = cat_1.merge(cat_2, on='CategoryID', suffixes=['_1', '_2'], how='outer', indicator=True)
cat_merge
```

Wypisanie wartości tylko z lewego zbioru

```python
filter = cat_merge['_merge'] == 'left_only'
cat_merge[filter]
```

Wypisanie wartości tylko z prawego zbioru

```python
filter = cat_merge['_merge'] == 'right_only'
cat_merge[filter]
```

Wypisanie wartości z obu zbiorów

```python
filter = cat_merge['_merge'] == 'both'
cat_merge[filter]
```

```python
cat_1.merge(cat_2, on='CategoryID', suffixes=['_1', '_2'], how='left', indicator=True)
```

```python
cat_1.merge(cat_2, on='CategoryID', suffixes=['_1', '_2'], how='right', indicator=True)
```
