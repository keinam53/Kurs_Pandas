### <a name='0'></a> Import bibliotek

```python
import pandas as pd
import plotly.express as px
```

### <a name='1'></a> Wygenerowanie danych

```python
from sklearn.datasets import make_blobs

data = make_blobs(n_samples=1000, centers=None, cluster_std=1, center_box=(-4, 4), random_state=42)[0]
df = pd.DataFrame(data, columns=['x1', 'x2'])
df.head()
```

### <a name='2'></a> Wizualizacja danych

```python
px.scatter(df, 'x1', 'x2', width=700, height=500, title='L;asteryzacja - Algorytm K-średnich')
```

### <a name='3'></a> Algorytm K-średnich

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
```

```python
y_kmeans = kmeans.predict(data)
y_kmeans[:10]
```

```python
df['y_kmeans'] = y_kmeans
df.head()
```

### <a name='4'></a> Wizualizacja klastrów

```python
px.scatter(df, 'x1', 'x2', color='y_kmeans', width=950, height=500, title='Algorytm K-średnich - 3 klastry')
```
