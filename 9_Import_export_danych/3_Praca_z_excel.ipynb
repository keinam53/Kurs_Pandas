```python
excel_writer = pd.ExcelWriter('cars.xlsx')
sheets['Old Europe'].to_excel(excel_writer, index=False, columns=['REGIONS/COUNTRIES', 2010],
                              sheet_name='Old Europe 2010')
sheets['Old Europe'].to_excel(excel_writer, index=False, columns=['REGIONS/COUNTRIES', 2011],
                              sheet_name='Old Europe 2011')
sheets['Old Europe'].to_excel(excel_writer, index=False, columns=['REGIONS/COUNTRIES', 2012],
                              sheet_name='Old Europe 2012')
excel_writer.save()
```
