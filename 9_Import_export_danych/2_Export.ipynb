**Export danych**

```python
import pandas as pd
import numpy as np
```

```python
df_blood = pd.DataFrame({'Group'     : ['0', '0', 'A', 'A', 'B', 'B', 'AB', 'AB'],
                        'Rh'         : ['+', '-', '+', '-', '+', '-', '+', '-'],
                        'Population' : [31,6,32,6,15,2,7,1]})
df_blood
```

Export danych do pliku csv

```python
df_blood.to_csv('blood.csv')
```

```python
df_blood_proc = df_blood['Population'] / 100
```

```python
df_blood_proc
```

```python
df_blood['Population'] = df_blood_proc
df_blood
```

Ponowny zapis do csv index=False - nie zapisze siÄ™ indeks

```python
df_blood.to_csv('blood.csv', index=False)
```

Eksport tylko wybranych kolumn

```python
df_blood.to_csv('blood.csv', index=False, columns=['Group', 'Rh'])
```

```python
pd.read_csv('blood.csv')
```

JeÅ›li przy zapisie dodajemy sep=... to taki sam musi byÄ‡ przy doczycie

```python
df_blood.to_csv('blood.csv', index=False, columns=['Group', 'Rh'], sep=';')
```

```python
pd.read_csv('blood.csv', sep=';')
```

mod='a' - dopisanie do istniejÄ…cego pliku ,header=False - nie wypisze siÄ™ kolejny nagÅ‚Ã³wek

```python
df_blood.to_csv('blood.csv', index=False, columns=['Group', 'Rh'], sep=':', mode='a', header=False)
```

```python
pd.read_csv('blood.csv', sep=';')
```
