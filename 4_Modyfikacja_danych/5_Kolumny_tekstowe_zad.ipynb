```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/Fortune_500_2017.csv',
                     usecols=['Rank','Title','Industry','Hqlocation','Sector'])
fortune.head()
```

```python
fortune['Sector'] = fortune['Sector'].str.upper()
fortune.head()
```

```python
fortune['Industry'] = fortune['Industry'].str.lower()
fortune[fortune['Industry'].str.contains('comp')].head()
```

```python
fortune[['City', 'State']] = fortune['Hqlocation'].str.split(',', expand=True)
fortune.head()
```

```python
def build_shortcut(row):
    industry = row['Industry']
    result = ''
    for letter in industry.split(' '):
        result += letter[0]
    return result
```

```python
build_shortcut({'Industry':'Factory Under Newspaper'})
```

```python
fortune['Industry_Short'] = fortune.apply(build_shortcut, axis=1)
fortune.head()
```
