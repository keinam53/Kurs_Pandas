Wczytanie tylko wybranych kolumn

```python
frame=pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/PublicTransitExpenses.csv',
                  usecols=['Agency', 'Reporter Type', 'Organization Type', 'Rail (Y/N)', 'Fixed Route (Y/N)',
                    'Service Costs', 'Tires and Tubes', 'Total Operating Expenses', 'Service Area Population'], 
                  low_memory=False)
frame.head()
```

```python
frame.info(memory_usage='deep')
```

Zmiana nazw tak żeby nie było nawiasów i spacji

```python
new_column_names = {'Agency': 'Agency',
                    'Reporter Type': 'ReporterType',
                    'Organization Type': 'OrgType',
                    'Service Area Population': 'Population',
                    'Rail (Y/N)': 'IsRail',
                    'Fixed Route (Y/N)': 'IsFixedRoute',
                    'Service Costs': 'Service Cost',
                    'Tires and Tubes': 'TiresTubesCost',
                    'Total Operating Expenses': 'TotalExpenses'
                    }
```

```python
frame.rename(columns=new_column_names, inplace=True)
frame.head()
```

Podstawowe sprawdzenie i zmiana typów

```python
frame['ReporterType'].nunique()
```

```python
frame['ReporterType'].count()
```

```python
frame['ReporterType'].value_counts()
```

```python
frame['ReporterType'] = frame['ReporterType'].astype('category')
frame.info(memory_usage='deep')
```

```python
frame['Agency'] = frame['Agency'].astype('category')
frame['OrgType'] = frame['OrgType'].astype('category')
frame.info(memory_usage='deep')
```

Wypełnienie wartości null

```python
frame['Population'].fillna(value=0, inplace=True)
frame[frame['Population'] > 0].head()
```

```python
frame.info(memory_usage='deep')
```

Zamiana wartości Y, N na True, False

```python
frame['IsRail'].replace(('Y', 'N'), (True, False), inplace=True)
frame['IsFixedRoute'].replace(('Y', 'N'), (True, False), inplace=True)
frame.head()
```

```python
frame.info(memory_usage='deep')
```

Jeśli w kolumnie logicznej jest null - zmiana na false

```python
frame['IsRail'].fillna(False, inplace=True)
frame['IsFixedRoute'].fillna(False, inplace=True)
frame['IsRail'] = frame['IsRail'].astype('bool')
frame['IsFixedRoute'] = frame['IsFixedRoute'].astype('bool')
frame.head()
```

```python
frame.info(memory_usage='deep')
```

Kolumny z kwotami - usunięcie $ i zamiana na float

```python
frame['Service Cost'] = frame['Service Cost'].str.replace('$', '')
frame['TiresTubesCost'] = frame['TiresTubesCost'].str.replace('$', '')
frame['TotalExpenses'] = frame['TotalExpenses'].str.replace('$', '')

frame['Service Cost'] = frame['Service Cost'].astype('float')
frame['TiresTubesCost'] = frame['TiresTubesCost'].astype('float')
frame['TotalExpenses'] = frame['TotalExpenses'].astype('float')
frame.head()
```

```python
frame.info(memory_usage='deep')
```
