Funkcja przypisująca komentarz ze względu na wysokość wydatków

```python
def get_comment(row):
    reporter_type = row['Reporter Type']
    cost = float(row['Total Operating Expenses'].replace('$', ''))
    if cost > 200000:
        comment = 'CLASS A'
    else:
        comment = 'CLASS B'
    return f'reporter_type / {comment}'
```

```python
frame.apply(get_comment, axis=1).head()
```
