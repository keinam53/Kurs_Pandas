**Przetwarzanie danych z obiektu grupy**

```python
import pandas as pd
import numpy as np
```

```python
products = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/WA_Sales_Products_2012-14.csv')
products.head()
```

```python
groups = products.groupby(by=['Retailer country'])
```

Wyświetlenie państw za pomocą pętli

```python
for country in groups:
    print(country[0])
```

```python
country_names = []
for country in groups:
    country_name.append(country[0])
country_names
```

Wyświetlenie nazwy kraju i ilości wierszy w tej grupie

```python
for country in groups:
    print(country[0], len(country[1]))
```

```python
for country, country_data in groups:
    print(f'Kraj: {country} Elementów: {len(country_data)}')
```

Różnica pomiędzy transakcją o największym i najmniejszym zysku

```python
for country, country_data in groups:
    print(country,
         country_data['Revenue'].max() - country_data["Revenue"].min())
```

Wyświetlenie kraju, maksymalnego zysku i jego indeksu

```python
for country, country_data in groups:
    print(country,
         country_data['Revenue'].max(),
         country_data['Revenue'].idxmax())
```

Wyświetlenie kraju, maksymalnego zysku i jego indeksu, wyświetlenie tego wiersza

```python
for country, country_data in groups:
    print(country,
         country_data['Revenue'].max(),
         country_data['Revenue'].idxmax(),
         country_data.loc[country_data['Revenue'].idxmax()])
```

Dodanie do DF największych wartości dla każdej grupy

```python
the_biggest = pd.DataFrame()
for country, country_data in groups:
    the_biggest = the_biggest.append(country_data.loc[country_data['Revenue'].idxmax])
the_biggest.head()
```

To samo używając nlargest

```python
the_biggest = pd.DataFrame()
for country, country_data in groups:
    the_biggest = the_biggest.append(country_data.nlargest(1, 'Revenue'))
the_biggest.head()
```
