```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import time
```

```python
df = pd.read_csv('C:/Users/Mariusz Baran/Data_Science_kurs/Data_science_pliki/course-files/marathon_results_2016.csv',
                usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'], index_col='Bib')
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
group_city = df.groupby(by='City')
```

```python
for city in group_city:
    print(city[0])
```

```python
for city in group_city:
    print(city[0], len(city))
```

```python
for city, city_data in group_city:
    print(city, len(city_data))
```

```python
for city, city_data in group_city:
    print(city, city_data.nsmallest(1, 'TotalSeconds'))
```

```python
the_best_city = pd.DataFrame()
for city, city_data in group_city:
    the_best_city = the_best_city.append(city_data.nsmallest(1, 'TotalSeconds'))
the_best_city.head()
```
